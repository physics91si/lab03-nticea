```python
# Don't edit this function
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample
```

```python
# Edit this function. "pass" tells Python to do nothing.
def parse_sample(words_only=False, sort_list=False):
    sample = load_sample()
    sample = sample.replace("--", " ")
    sample = sample.replace("-", " ")
    words = sample.split()  
    if(words_only):
        exclude = set(['!','@','#','$','%','^','&','*','(',')','\"','?','.','\'','[',']', ',',';', ':']);
        for i in range(len(words)):
            word = ''.join(c for c in words[i] if c not in exclude)
            word = word.lower()
            words[i] = word
    if(sort_list):
        (words).sort()
    return words
```

```python
import math

def count_freq(lst):
    lst_set = set(lst)
    freq_dict = {}
    for w in lst_set:
        freq_dict[w] = 0
    for w in lst:
        freq_dict[w] += 1
    return freq_dict

def mean(lst):
    s = 0
    for x in lst:
        s += x
    return (s/len(lst))

def stdev(lst):
    avg = mean(lst)
    s = 0;
    for x in lst:
        s += (x - avg)**2
    return math.sqrt(s/(len(lst) - 1))

def median(lst):
    lst.sort()
    return lst[(len(lst))//2]
```

```python
def print_stats(lst):
    print("The frequency distribution of words in Hamlet is:", count_freq(parse_sample(True,True)))
    print("The average number of times a word appears is:", mean(list(count_freq(parse_sample(True,True)).values())))
```
